# 2. Conceitos fundamentais

## 2.1 Variável aleatória


<div align="justify">

Uma variável aleatória é uma função que associa a cada resultado de um experimento aleatório um valor numérico. Essas variáveis podem ser classificadas em dois tipos principais: discretas e contínuas (Devore, 2006).

As variáveis aleatórias discretas assumem valores finitos ou enumeráveis, sendo comuns em experimentos que envolvem contagem, como o número de falhas em um conjunto de corpos de prova. Já as variáveis contínuas podem assumir qualquer valor em um intervalo, sendo típicas em medições físicas, como resistência, deformação e deslocamento.

No contexto da engenharia de materiais, a resistência à compressão ou à flexão de um compósito cimentício é um exemplo clássico de variável contínua, enquanto o número de fissuras em um elemento estrutural pode ser tratado como uma variável discreta.


</div>

## 2.2 Distribuição de probabilidade





<div align="justify">

A distribuição de probabilidade descreve como os valores de uma variável aleatória estão distribuídos, associando probabilidades a cada possível resultado. No caso de variáveis discretas, essa distribuição é definida por uma função de probabilidade. Para variáveis contínuas, é descrita por uma função densidade de probabilidade (Montgomery e Runger, 2014; Bittencourt, 2014).

As distribuições de probabilidade permitem não apenas descrever o comportamento dos dados, mas também realizar inferências e previsões, sendo fundamentais no planejamento e análise de experimentos.

</div>

#  3. Distribuição Binomial

---



<div align="justify">

A distribuição binomial é uma das principais distribuições de probabilidade discreta, sendo utilizada para modelar experimentos compostos por um número fixo de ensaios independentes, nos quais cada ensaio apresenta apenas dois resultados possíveis: sucesso ou fracasso.

Segundo Devore (2006), a distribuição binomial descreve o número de sucessos em uma sequência de experimentos independentes com probabilidade constante de ocorrência. Essa estrutura é amplamente utilizada em problemas de engenharia que envolvem decisões binárias, como aprovação ou falha de elementos estruturais.

o material apresentado em aula, esse conceito foi ilustrado por meio do lançamento de moedas, onde cada lançamento representa um ensaio e os resultados possíveis são “cara” (sucesso) ou “coroa” (fracasso) . Esse exemplo pode ser diretamente extrapolado para diversas aplicações na engenharia.

</div>


<div style="background-color:#f9f0ff; padding:15px; border-left:5px solid #722ed1">

**Definição:**

Uma variável aleatória \( X \) segue uma distribuição binomial quando representa o número de sucessos em \( n \) ensaios independentes, cada um com probabilidade constante de sucesso \( p \).

</div>



###  Função de probabilidade

$$
P(X = k) = \binom{n}{k} p^k (1 - p)^{n - k}
$$

<div align="justify">

Onde:

- \( n \): número total de ensaios  
- \( k \): número de sucessos desejados  
- \( p \): probabilidade de sucesso  
- \( (1 - p) \): probabilidade de fracasso  


</div>


<div align="justify">

A distribuição binomial permite avaliar a probabilidade de ocorrência de um determinado número de eventos em um conjunto finito de experimentos. Na engenharia civil, isso é particularmente útil na análise de fenômenos discretos, como a ocorrência de falhas em corpos de prova, presença de fissuras ou conformidade de elementos estruturais com critérios normativos.

</div>

<div style="background-color:#e6f7ff; padding:15px; border-left:5px solid #1890ff">

**Exemplo 1 (Engenharia de materiais):**

Considere um conjunto de 10 corpos de prova de compósito cimentício reforçado com fibras vegetais. Suponha que a probabilidade de um corpo de prova apresentar fissuração precoce seja de 20% \( p = 0,2 \).

Deseja-se determinar a probabilidade de exatamente 3 corpos de prova apresentarem fissuração precoce.

Parâmetros do problema:

- \( n = 10 \)  
- \( k = 3 \)  
- \( p = 0,2 \)  

Esse tipo de abordagem permite avaliar a confiabilidade do material e compreender a variabilidade do comportamento experimental.

</div>

<div style="background-color:#fffbe6; padding:15px; border-left:5px solid #faad14">

**Observação:**

A aplicação da distribuição binomial pressupõe que os ensaios sejam independentes e que a probabilidade de sucesso permaneça constante. Em materiais cimentícios, essas condições podem ser influenciadas por fatores como variabilidade na matriz, distribuição de fibras e condições de moldagem.

</div>

### Implementação computacional

<div align="justify">

Para complementar a análise teórica, é possível utilizar ferramentas computacionais para calcular probabilidades e visualizar o comportamento da distribuição binomial.

Neste exemplo, será considerada a situação previamente descrita, envolvendo corpos de prova de compósitos cimentícios reforçados com fibras. O objetivo é calcular a probabilidade de ocorrência de diferentes números de falhas (fissuração precoce) e representar graficamente essa distribuição.

A utilização de linguagens de programação, como Python, permite automatizar cálculos e gerar visualizações que facilitam a interpretação dos resultados, sendo uma ferramenta amplamente empregada no planejamento e análise de experimentos.

</div>

```python
# Importação das bibliotecas
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Parâmetros do problema
n = 10
p = 0.2

# Valores possíveis
k = np.arange(0, n+1)

# Probabilidades
probabilidades = binom.pmf(k, n, p)

# Mostrar valores
for i in range(len(k)):
    print(f"P(X = {k[i]}) = {probabilidades[i]:.4f}")

# Gráfico
plt.figure()
plt.bar(k, probabilidades)
plt.xlabel("Número de falhas")
plt.ylabel("Probabilidade")
plt.title("Distribuição Binomial")
plt.show()
```

### Interpretação dos resultados

<div align="justify">

O gráfico apresentado ilustra a distribuição de probabilidade do número de corpos de prova que podem apresentar fissuração precoce.

Observa-se que os valores mais prováveis concentram-se em baixos números de falhas, o que é esperado considerando a probabilidade de falha adotada \( p = 0,2 \). À medida que o número de falhas aumenta, a probabilidade associada diminui significativamente.

Esse comportamento é coerente com sistemas em que a ocorrência de falhas individuais possui baixa probabilidade, sendo a distribuição binomial uma ferramenta adequada para representar esse tipo de fenômeno.

Na engenharia de materiais, essa análise permite avaliar a confiabilidade de compósitos cimentícios e compreender a variabilidade dos resultados experimentais, auxiliando na tomada de decisões quanto ao desempenho e segurança dos materiais.

</div>

#  4. Distribuição Normal

---



<div align="justify">

A distribuição normal é uma das mais importantes distribuições de probabilidade contínua, sendo amplamente utilizada na modelagem de fenômenos naturais e experimentais. Sua relevância decorre do fato de que diversos processos físicos e propriedades de materiais apresentam comportamento aproximadamente normal, especialmente quando influenciados por múltiplos fatores independentes.

Segundo Devore (2016), a distribuição normal é frequentemente empregada para representar variáveis contínuas, como resistência mecânica, deformações e propriedades físicas de materiais, devido à sua capacidade de descrever adequadamente a variabilidade experimental.

No contexto apresentado em aula, observou-se que, à medida que o número de ensaios aumenta, distribuições discretas, como a binomial, tendem a se aproximar da distribuição normal, evidenciando sua importância teórica e prática.

</div>

<div style="background-color:#f9f0ff; padding:15px; border-left:5px solid #722ed1">

**Definição:**

Uma variável aleatória contínua segue uma distribuição normal quando sua função densidade de probabilidade apresenta formato simétrico em torno da média, sendo caracterizada pelos parâmetros média $( \mu )$ e desvio padrão $( \sigma )$.

</div>

###  Função densidade de probabilidade

$$
f(x) = \frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{(x - \mu)^2}{2\sigma^2}}
$$

<div align="justify">

Na expressão da função densidade, a média $( \mu )$ define a posição central da distribuição, enquanto o desvio padrão $( \sigma )$ controla a dispersão dos valores em torno dessa média. Assim, quanto maior o desvio padrão, mais espalhados estarão os dados. Quanto menor, mais concentrados eles estarão próximos ao valor médio.

</div>

<div align="center">

**Figura 1 –** Influência do desvio padrão na forma da distribuição normal. Fonte: Devore (2006).

</div>


<div align="justify">

Observa-se na Figura 1 que, mantendo a média constante, o aumento do desvio padrão resulta em maior dispersão dos dados, enquanto valores menores de $ \sigma $ produzem distribuições mais concentradas.

</div>

### Propriedades da distribuição normal

<div align="justify">

A distribuição normal apresenta propriedades fundamentais que a tornam amplamente aplicável:

- É simétrica em torno da média;
- A média, a mediana e a moda coincidem;
- A área total sob a curva é igual a 1;
- A dispersão dos dados é controlada pelo desvio padrão;
- Valores mais distantes da média apresentam menor probabilidade de ocorrência.

Além disso, conforme discutido em aula, as probabilidades associadas à distribuição normal correspondem às áreas sob a curva, sendo possível determinar a probabilidade de ocorrência de determinados intervalos de valores.

</div>

###  Normal reduzida (Padronização)

<div align="justify">

Para facilitar o cálculo de probabilidades, utiliza-se a transformação da variável original \( X \) em uma variável padronizada \( Z \). Essa transformação converte qualquer distribuição normal em uma distribuição normal padrão, cuja média é igual a 0 e o desvio padrão é igual a 1.

</div>

$$
Z = \frac{X - \mu}{\sigma}
$$

### Cálculo de probabilidades

<div align="justify">

Na distribuição normal, as probabilidades são obtidas a partir da área sob a curva. Para isso, utiliza-se a função distribuição acumulada, que representa a probabilidade da variável aleatória assumir valores menores ou iguais a um determinado valor.

Em termos gerais, essa probabilidade pode ser representada por:

$$
P(X \leq x)
$$

A partir dessa lógica, diferentes situações podem ser analisadas.

</div>

### Formas usuais de cálculo

<div align="justify">

As probabilidades mais frequentes na distribuição normal são:

- Probabilidade acumulada à esquerda:

$$
P(X < a)
$$

- Probabilidade à direita:

$$
P(X > a) = 1 - P(X < a)
$$

- Probabilidade entre dois valores:

$$
P(a < X < b) = P(X < b) - P(X < a)
$$


</div>

<div align="justify">

A Figura 2 ilustra o cálculo da probabilidade entre dois valores como a diferença entre áreas acumuladas, evidenciando a interpretação geométrica da distribuição normal.

</div>

Essas relações são particularmente úteis na análise de resultados experimentais, permitindo quantificar a probabilidade de um material apresentar desempenho inferior, superior ou dentro de um intervalo de interesse.

</div>

###  Regra empírica da distribuição normal

<div align="justify">

A distribuição normal apresenta uma propriedade importante conhecida como regra empírica, segundo a qual:

- Aproximadamente 68% dos valores encontram-se no intervalo $( \mu \pm 1\sigma )$;
- Aproximadamente 95% dos valores encontram-se no intervalo $( \mu \pm 2\sigma )$;
- Aproximadamente 99,7% dos valores encontram-se no intervalo $( \mu \pm 3\sigma )$.

Essa característica é extremamente útil na interpretação da variabilidade experimental, permitindo avaliar o quão dispersos estão os resultados obtidos em ensaios laboratoriais.

</div>

<div style="background-color:#fffbe6; padding:15px; border-left:5px solid #faad14">
<div align="justify">

**Aplicação na engenharia de materiais:**

Na análise de materiais cimentícios reforçados com fibras, a distribuição normal pode ser empregada para representar propriedades contínuas, como resistência à compressão, resistência à flexão, absorção de água, módulo de elasticidade e outras respostas experimentais. Nesses casos, a média representa o comportamento central do material, enquanto o desvio padrão expressa a dispersão dos resultados, permitindo avaliar sua uniformidade e confiabilidade.

</div>

<div style="background-color:#e6f7ff; padding:15px; border-left:5px solid #1890ff">

**Exemplo (Engenharia de materiais):**

Considere os resultados de resistência à flexão de corpos de prova de compósito cimentício reforçado com fibras vegetais.

Suponha que os resultados apresentem:

- média $ \mu = 12 $ MPa  
- desvio padrão $ \sigma = 2 $ MPa  

Deseja-se determinar a probabilidade de um corpo de prova apresentar resistência inferior a 10 MPa.

Esse tipo de análise permite avaliar a confiabilidade do material e identificar a probabilidade de ocorrência de resultados abaixo de um limite crítico.

</div>

### Resolução do exemplo 1
<div align="justify">

Aplicando a transformação para a variável padronizada:

$$
Z = \frac{X - \mu}{\sigma}
$$

Substituindo os valores do problema:

$$
Z = \frac{10 - 12}{2} = -1
$$

Portanto, o problema passa a ser determinar:

$$
P(X < 10) = P(Z < -1)
$$

Pela distribuição normal padrão, essa probabilidade é aproximadamente igual a 0,1587, ou seja, cerca de 15,87%.

</div>

### Implementação computacional

<div align="justify">

Além da resolução analítica por padronização, a distribuição normal pode ser analisada por meio de ferramentas computacionais. O uso de Python permite calcular probabilidades de forma rápida e representar graficamente o comportamento da distribuição, facilitando a interpretação dos resultados experimentais.

No exemplo a seguir, será calculada a probabilidade de um corpo de prova apresentar resistência à flexão inferior a 10 MPa, considerando média de 12 MPa e desvio padrão de 2 MPa.

</div>

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parâmetros do problema
mu = 12      # média
sigma = 2    # desvio padrão

# Intervalo de valores para o gráfico
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 400)

# Função densidade da distribuição normal
y = norm.pdf(x, mu, sigma)

# Probabilidade de resistência inferior a 10 MPa
prob = norm.cdf(10, mu, sigma)

print(f"Probabilidade de resistência < 10 MPa: {prob:.4f}")

# Gráfico
plt.figure()
plt.plot(x, y, label="Distribuição Normal")
plt.xlabel("Resistência à flexão (MPa)")
plt.ylabel("Densidade de probabilidade")
plt.title("Distribuição Normal - Resistência à Flexão")

# Área sombreada
x_fill = np.linspace(mu - 4*sigma, 10, 200)
y_fill = norm.pdf(x_fill, mu, sigma)
plt.fill_between(x_fill, y_fill, alpha=0.5, label="P(X < 10 MPa)")

plt.legend()
plt.show()
```

### Interpretação dos resultados

<div align="justify">

O resultado indica que a probabilidade de um corpo de prova apresentar resistência à flexão inferior a 10 MPa é de aproximadamente 15,87%. Graficamente, essa probabilidade corresponde à área sob a curva normal à esquerda desse valor.

Na prática, esse tipo de análise é extremamente útil no controle tecnológico de materiais, pois permite quantificar o risco de obtenção de resultados abaixo de um limite de desempenho considerado aceitável. Em materiais cimentícios reforçados com fibras, isso contribui para a avaliação da uniformidade do processo produtivo e da confiabilidade mecânica do compósito.

</div>

P(Z > 1,25) \approx 0,1056

# 5. Distribuição de Poisson

---



<div align="justify">

A distribuição de Poisson é uma distribuição de probabilidade discreta utilizada para modelar o número de ocorrências de um evento em um intervalo fixo de tempo, área, volume ou extensão, desde que esses eventos ocorram de forma independente e com uma taxa média aproximadamente constante.

Essa distribuição é empregada para representar a quantidade de eventos em um determinado intervalo, sendo especialmente útil em problemas de contagem. Essa distribuição é particularmente relevante quando se deseja modelar fenômenos relativamente raros, como defeitos, falhas localizadas, poros ou fissuras em materiais.

No contexto da engenharia de materiais, sua aplicação é bastante conveniente para descrever situações em que o interesse está centrado no número de ocorrências de um determinado evento por unidade de área, por corpo de prova ou por elemento analisado.

</div>

<div style="background-color:#f9f0ff; padding:15px; border-left:5px solid #722ed1">

**Definição:**

Uma variável aleatória \( X \) segue uma distribuição de Poisson quando representa o número de vezes que um evento ocorre em um intervalo fixo, sendo esse comportamento descrito por uma taxa média de ocorrência $( \lambda )$.

</div>

### Função de probabilidade

$$
P(X = k) = \frac{e^{-\lambda}\lambda^k}{k!}
$$

<div align="justify">

Onde:

- \( k \) = número de ocorrências do evento;
- $( \lambda )$ = número médio de ocorrências no intervalo considerado;
- \( e \) = base do logaritmo natural.

Na distribuição de Poisson, a média e a variância são ambas iguais a $( \lambda )$, propriedade clássica dessa distribuição e muito útil na interpretação de dados de contagem.

</div>

### Condições de aplicação

<div align="justify">

A distribuição de Poisson é adequada quando:

- Os eventos são contados em um intervalo fixo;
- As ocorrências são independentes;
- A taxa média de ocorrência permanece aproximadamente constante;
- A probabilidade de ocorrência simultânea de muitos eventos em um intervalo muito pequeno é baixa.

Além disso, a distribuição de Poisson pode ser entendida como um caso limite da distribuição binomial quando o número de ensaios é grande, a probabilidade de sucesso em cada ensaio é pequena e o produto \( np \) permanece constante.

</div>

<div style="background-color:#fffbe6; padding:15px; border-left:5px solid #faad14">

**Aplicação na engenharia de materiais:**

Na engenharia de materiais, a distribuição de Poisson pode ser utilizada para modelar o número de fissuras observadas em uma superfície, a quantidade de defeitos em um lote de amostras, o número de poros visíveis em uma seção analisada ou a ocorrência de falhas localizadas em compósitos cimentícios reforçados com fibras.

Esse tipo de abordagem é especialmente útil quando o interesse não está no valor contínuo de uma propriedade mecânica, mas sim na contagem de eventos discretos associados ao comportamento do material.

</div>

<div style="background-color:#e6f7ff; padding:15px; border-left:5px solid #1890ff">

**Exemplo 1 (Fissuras em material cimentício):**

Suponha que, em uma determinada análise experimental, seja observado que uma placa de compósito cimentício reforçado com fibras apresenta, em média, 2 fissuras por área inspecionada.

Deseja-se determinar a probabilidade de serem observadas exatamente 3 fissuras nessa mesma área.

Nesse caso:

- $\lambda = 2 $
- $\ k = 3 \$

</div>

### Resolução do exemplo 1

<div align="justify">

Aplicando a função de probabilidade da distribuição de Poisson:

$$
P(X = k) = \frac{e^{-\lambda}\lambda^k}{k!}
$$

Substituindo os valores:

$$
P(X = 3) = \frac{e^{-2}2^3}{3!}
$$

$$
P(X = 3) = \frac{e^{-2}\cdot 8}{6}
$$

$$
P(X = 3) \approx 0,1804
$$

Portanto, a probabilidade de observar exatamente 3 fissuras na área analisada é de aproximadamente 18,04%.

</div>

<div style="background-color:#e6f7ff; padding:15px; border-left:5px solid #1890ff">

**Exemplo 2 (Defeitos em corpos de prova):**

Considere que, em um processo de moldagem de corpos de prova de argamassa reforçada com fibras vegetais, o número médio de defeitos superficiais observados por amostra seja igual a 1,5.

Deseja-se determinar a probabilidade de uma amostra apresentar no máximo 1 defeito superficial.

Nesse caso:

- $ \lambda = 1,5 $
- deseja-se calcular \$( P(X \leq 1) \)$

</div>

### Resolução do exemplo 2

<div align="justify">

A probabilidade de uma amostra apresentar no máximo 1 defeito é dada por:

$$
P(X \leq 1) = P(X = 0) + P(X = 1)
$$

Calculando cada parcela:

$$
P(X = 0) = \frac{e^{-1,5}(1,5)^0}{0!} = e^{-1,5}
$$

$$
P(X = 1) = \frac{e^{-1,5}(1,5)^1}{1!} = 1,5e^{-1,5}
$$

Logo:

$$
P(X \leq 1) = e^{-1,5} + 1,5e^{-1,5}
$$

$$
P(X \leq 1) \approx 0,5578
$$

Assim, a probabilidade de uma amostra apresentar no máximo 1 defeito superficial é de aproximadamente 55,78%.

</div>

### Implementação computacional

<div align="justify">

A implementação computacional permite calcular rapidamente as probabilidades associadas à distribuição de Poisson e representar graficamente o comportamento da variável aleatória.

No exemplo a seguir, considera-se o caso em que o número médio de fissuras por área inspecionada é igual a 2. O objetivo é calcular as probabilidades para diferentes números de fissuras e visualizar a distribuição por meio de um gráfico de barras.


</div>

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Parâmetro da distribuição
lam = 2  # número médio de fissuras

# Valores possíveis de ocorrência
k = np.arange(0, 11)

# Probabilidades
probabilidades = poisson.pmf(k, lam)

# Exibir probabilidades
for i in range(len(k)):
    print(f"P(X = {k[i]}) = {probabilidades[i]:.4f}")

# Probabilidade de exatamente 3 fissuras
prob_3 = poisson.pmf(3, lam)
print(f"\nProbabilidade de exatamente 3 fissuras: {prob_3:.4f}")

# Gráfico
plt.figure()
plt.bar(k, probabilidades)
plt.xlabel("Número de fissuras")
plt.ylabel("Probabilidade")
plt.title("Distribuição de Poisson - Número de fissuras")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

lam = 2
k = np.arange(0, 11)
probabilidades = poisson.pmf(k, lam)

plt.figure()
bars = plt.bar(k, probabilidades)

# Destacar a barra referente a k = 3
bars[3].set_alpha(0.8)

plt.xlabel("Número de fissuras")
plt.ylabel("Probabilidade")
plt.title("Distribuição de Poisson - Destaque para P(X = 3)")
plt.show()
```

### Interpretação dos resultados

<div align="justify">

O gráfico evidencia o comportamento típico da distribuição de Poisson para contagens discretas. Observa-se que, para uma taxa média de 2 fissuras por área, os maiores valores de probabilidade concentram-se em pequenas contagens, enquanto números maiores de fissuras tornam-se progressivamente menos prováveis.

Na prática, esse tipo de análise é útil para avaliar a frequência esperada de defeitos ou fissuras em materiais cimentícios, contribuindo para o controle de qualidade e para a interpretação do comportamento experimental de compósitos reforçados com fibras.

# 6. Comparação entre distribuições

---




<div align="justify">

As distribuições binomial, normal e de Poisson apresentam características distintas, sendo aplicáveis a diferentes tipos de problemas na engenharia.
De forma geral:

- A distribuição binomial é utilizada para modelar eventos discretos com dois resultados possíveis (sucesso ou fracasso);
- A distribuição normal é aplicada a variáveis contínuas, especialmente em fenômenos naturais e propriedades de materiais;
- A distribuição de Poisson é empregada na modelagem de eventos raros, associados à contagem de ocorrências em intervalos definidos.

Além disso, existe uma relação importante entre essas distribuições: a distribuição de Poisson pode ser interpretada como uma aproximação da distribuição binomial em situações em que o número de ensaios é elevado e a probabilidade de ocorrência é pequena, enquanto a distribuição normal pode aproximar a binomial quando o número de ensaios é suficientemente grande.
